```python
# JOINED KINGDOM, and sort based on microbiome cohort
Endo_data_micro = pd.read_csv("Endo_microbiome_data_joinedkingdom.csv") 
Endo_data_micro
```

```python
# split the table controls and patients
# select only rows with controls
controls = Endo_data_micro[Endo_data_micro['Stage 1-2'] == 1]
controls
```

```python
# select only rows with patients
patients = Endo_data_micro[Endo_data_micro['Stage 1-2'] == 2]
patients
```

```python
# select bacteria which are not in controls
controls_no_bacteria = controls.loc[:, (controls == 0).all()]
controls_no_bacteria
```

```python
# list control columns
list_controls = controls_no_bacteria.columns.values.tolist() 
list_controls
```

```python
# select bacteria which are in patients
patient_bacteria = patients.loc[:, (patients != 0).any()]
patient_bacteria
```

```python
# list of patient columns
list_patients = patient_bacteria.columns.values.tolist() 
list_patients
```

```python
# list bacteria in both lists
list(set(list_controls).intersection(list_patients))
```
